# Decision Tree (edible vs poisonous) â€” Wine Quality

Notebook ini berisi langkah-langkah EDA singkat, training pipeline, tuning ccp_alpha (cost-complexity pruning), evaluasi, dan visualisasi. Untuk menjalankan end-to-end secara otomatis gunakan `decisiontree_generic.py`.


```python
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import GridSearchCV, StratifiedKFold
import matplotlib.pyplot as plt
import joblib
import json
```

## Load data

```python
df = pd.read_csv('winequalityN.csv')
df.shape
```

## Quick EDA

```python
df.head()
```

```python
df.isna().sum()
```

## Create binary label: edible (quality >= 6) vs poisonous

Kita pilih threshold 6 (umum dipakai pada dataset winequality).

```python
y = (df['quality'].astype(float) >= 6).astype(int)
X = df.drop(columns=['quality'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape
```

## Build preprocessor (categorical -> OneHotEncoder(handle_unknown='ignore'), numeric -> median imputer)

```python
categorical_cols = X_train.select_dtypes(include=['object','category']).columns.tolist()
numeric_cols = [c for c in X_train.columns if c not in categorical_cols]
cat_pipe = Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), ('onehot', OneHotEncoder(handle_unknown='ignore', sparse=False))])
num_pipe = Pipeline([('imputer', SimpleImputer(strategy='median'))])
preprocessor = ColumnTransformer([('cat', cat_pipe, categorical_cols), ('num', num_pipe, numeric_cols)], remainder='drop', verbose_feature_names_out=False)
```

## Fit preprocessor and find candidate ccp_alphas

```python
preprocessor.fit(X_train)
X_tr = preprocessor.transform(X_train)
base = DecisionTreeClassifier(random_state=42)
base.fit(X_tr, y_train)
path = base.cost_complexity_pruning_path(X_tr, y_train)
alphas = np.unique(path.ccp_alphas)
alphas = alphas[:-1] if alphas.size>1 else alphas
alphas[:10]
```

## GridSearchCV over ccp_alpha (5-fold stratified CV)

```python
pipe = Pipeline([('preprocessor', preprocessor), ('clf', DecisionTreeClassifier(random_state=42))])
param_grid = {'clf__ccp_alpha': np.unique(np.concatenate(([0.0], alphas))).tolist()}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
gs = GridSearchCV(pipe, param_grid=param_grid, cv=cv, scoring='f1', n_jobs=-1, verbose=1)
gs.fit(X_train, y_train)
gs.best_params_, gs.best_score_
```

## Evaluate best model and baseline (no pruning)

```python
best = gs.best_estimator_
baseline = Pipeline([('preprocessor', preprocessor), ('clf', DecisionTreeClassifier(random_state=42))])
baseline.fit(X_train, y_train)
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
X_test_tr = best.named_steps['preprocessor'].transform(X_test)
y_pred_best = best.named_steps['clf'].predict(X_test_tr)
y_pred_base = baseline.named_steps['clf'].predict(X_test_tr)
print('Baseline acc', accuracy_score(y_test, y_pred_base))
print('Best acc', accuracy_score(y_test, y_pred_best))
print('\nClassification report (best)')
print(classification_report(y_test, y_pred_best))
```

## Feature importances (after preprocessing)
Kita extract nama fitur dari OneHotEncoder lalu tampilkan top features.

```python
cat_names = best.named_steps['preprocessor'].named_transformers_['cat'].named_steps['onehot'].get_feature_names_out(categorical_cols).tolist() if categorical_cols else []
feat_names = cat_names + numeric_cols
fi = best.named_steps['clf'].feature_importances_
imp = sorted(zip(feat_names, fi), key=lambda x: x[1], reverse=True)[:20]
imp
```

## Precision-Recall curve

```python
from sklearn.metrics import precision_recall_curve, average_precision_score
y_score = best.named_steps['clf'].predict_proba(X_test_tr)[:,1]
precision, recall, _ = precision_recall_curve(y_test, y_score)
ap = average_precision_score(y_test, y_score)
plt.figure(figsize=(6,6))
plt.plot(recall, precision, label=f'AP={ap:.3f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall curve')
plt.legend()
plt.grid()
plt.show()
```

## Visualize pruned tree (best)
Jika terlalu besar, gunakan `max_depth`/atau gunakan node limit yang diinginkan saat tuning.

```python
plt.figure(figsize=(20,12))
plot_tree(best.named_steps['clf'], feature_names=feat_names, filled=True, impurity=True, fontsize=8)
plt.show()
```

## Save artifacts (models + preprocessor + reports)
Simpan pipeline ke folder `models/` dan metrics/plots ke `reports/`

```python
Path('models').mkdir(exist_ok=True)
joblib.dump(best, 'models/tree_pipeline.joblib')
joblib.dump(preprocessor, 'models/preprocessor.joblib')
print('Saved models to models/')
```
